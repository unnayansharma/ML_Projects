1. Generate 20 data points from function f(x) = cos(2πx) + x*2π + noise where noise∼N (0, 0.004) with x ranging from 0 to 2π.

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
def makedata(size):
  y=[]
  X=[]
  data=[]
  for i in range(size):
    a=2*math.pi
    x=np.random.uniform(0,a)
    X.append(x)
    noise=np.random.normal(0,0.004)
    FX=math.cos(a*x) + (x/a) + noise
    y.append(FX)
    data.append((x,FX))
  df=pd.DataFrame(data,columns=['X','FX'])
  return df,X,y
```

2. Fit a polynomial regression with optimal weight vector w∗ and plot the curves for dif-ferent degree of polynomials M = 1, 2, 3, 5, 7, 10. Explain your observations by plotting the data points generated and the curve obtained for different values of M.

```python
def PolynomialRegression(degree,size):
  phi=np.array([0]*(size*degree),dtype=float)
  phi=phi.reshape(degree,size)
  for i in range(degree):
    for j in range(size):
      phi[i][j]=math.pow(X[j],i)
  phi=np.transpose(phi)
  w=np.dot(np.transpose(phi),phi)
  w=np.linalg.inv(w)
  w=np.dot(w,np.transpose(phi))
  w=np.dot(w,y)
  pred_x=np.array([0]*size,dtype=float)
  for k in range(size):
    for i in range(degree):
      pred_x[k]+=(w[i]*(X[k]**i))
  return pred_x
```

```python
degree=1
size=20
df,X,y=makedata(size)
pred_x=PolynomialRegression(degree,size)
plt.scatter(X,y)
plt.scatter(X,pred_x,color="red")
plt.show()
```
